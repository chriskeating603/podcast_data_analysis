###Setting Up the Environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
%load_ext jupyternotify
# import nltk
```

###Getting the Data

```python
# data source is https://www.kaggle.com/thoughtvector/podcastreviews
df_podcasts = pd.read_csv('./data/podcasts.csv')
df_cat = pd.read_csv('./data/categories.csv')
```

```python
# select the full dataset for more robust analysis, or select first 1k rows for testing analysis
df_reviews = pd.read_csv('./data/reviews.csv')
# df_reviews = pd.read_csv('./data/reviews.csv')[:1000]
```

###Transforming the data

```python
# Reduce data to last 18 months
df_reviews['created_at'] = df_reviews['created_at'].apply(lambda x: datetime.fromisoformat(str(x)))
cutoff = datetime.fromisoformat('2019-05-31T12:00:00-07:00')
df_reviews = df_reviews[(df_reviews['created_at'] > cutoff)]

# Convert to monthly
df_reviews['created_at'] = df_reviews['created_at'].apply(lambda x: datetime.strftime(x, '%Y-%m'))
```

```python
# Reduce our dataset to the 12 most reviewed podcasts in the dataset
counts = df_reviews['podcast_id'].value_counts()
df_reviews['num_reviews'] = df_reviews['podcast_id'].apply(lambda x: counts[x])
df_reviews = df_reviews[(df_reviews['num_reviews'] > 2500 )]
```

```python
# Prepare table for avg scores for each podcast over the last 18 months
pt = pd.pivot_table(df_reviews, values=['rating'], index=['podcast_id'], columns=['created_at'], aggfunc=[np.mean])
pt.columns = pt.columns.get_level_values(2) #reduce multi-index to single index
```

```python
# Prepare table to join avg scores and podcast metadata
df_podcasts.set_index('podcast_id', inplace=True)
```

```python
# Merge tables to create table with podcast names and avg scores over past 18 months
df_avg_rating_by_month = pd.merge(pt, df_podcasts[['title']],left_index=True, right_index=True,how='left')

# Display full table
pd.options.display.max_columns = None
df_avg_rating_by_month.dropna(subset=['title'], inplace=True)
display(df_avg_rating_by_month)
```

###Visualizing the Data

```python
# Prepare table for visualization
df_avg_rating_by_month = df_avg_rating_by_month.set_index('title', drop=True).transpose()
```

```python
# Look at data
df_avg_rating_by_month.head()
```

```python
# Display all shows on one chart
ax = df_avg_rating_by_month.plot(colormap='Dark2', figsize=(15,10), linewidth=3, fontsize=12)
ax.set_xlabel('Date')
ax.set_ylabel('Avg Rating')
plt.show()
```
